```python
import numpy as np
import pandas as pd
```

```python
# Load rural/urban population dataset from US Census
# data souce: https://www.census.gov/data/datasets/2017/demo/popest/counties-total.html
county_pop = pd.read_csv('data.csv', sep=';',encoding='latin-1')
county_pop
```

```python
# finding pairs of state and their abbreviation
state_abb = pd.read_csv('states.csv', encoding='latin-1')
state_abb    
```

```python
# Construct a dictonary such that keys are those states and values are the abbreviation of each state.
state = {}
for index, row in state_abb.iterrows():
    state[row[0]] = row[1]
state
```

```python
# for each county, add the state which this county belongs to
STATE = []
for index, row in county_pop.iterrows():
    STATE.append(state[row[1]])
county_pop['STATE'] = STATE

# create a column 'state county', i.e. a string contains the name of a state and the name of a county 
# which belongs to such state. because there exists counties with the same name but in differenct states
statecounty = []

# delete ‘Conuty’ in city names, so that all row data are splitting into lists of the same length.
for index, row in county_pop.iterrows():
    if row[0].split(' ')[-1] == 'County':
        statecounty.append(row[5] + ' ' + ' '.join(row[0].split(' ')[:-1]))
    else:
        statecounty.append(row[5] + ' ' + ' '.join(row[0].split(' ')))

# append statecounty data into the data set.
county_pop['statecounty'] = statecounty

# calculate the percentage of rural population out of total county population
county_pop['ruralpercent'] = county_pop['2010 Census Rural Population']/county_pop['2010 Census Total Population']
# classify a conuty as rural if its rural population > urban population
# otherwise, classify it as urban
ruralurban = []
for index, row in county_pop.iterrows():
    if row['ruralpercent'] >= 0.5:
        ruralurban.append(1)
    else:
        ruralurban.append(0)
# append the label of each county into the data set.
county_pop['ruralurban'] = ruralurban
county_pop
```

```python
county_pop[["statecounty", 'ruralurban']].to_csv('county_pop2.csv')
```
